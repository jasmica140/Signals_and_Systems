```python
#2.1
import os 
import librosa

x, fs = librosa.load(os.path.join('Data-20221031','imperfect_sos_cw_700_15_wpm.mp3'))

import IPython.display as ipd 
ipd.display(ipd.Audio(data=x, rate=fs))

import matplotlib.pyplot as plt
```

```python
#2.2
plt.figure()
plt.stem(x,markerfmt='b.')
plt.xlabel('Sample')
plt.ylabel('Value')
plt.title('Clean SOS')
      
plt.show()
```

```python
#2.3
plt.figure()
t = range(1000,1200) 
plt.stem(t,x[t],markerfmt='b.') 
plt.xlabel('Sample') 
plt.ylabel('Value') 
plt.title('Clean SOS')
plt.show()
```
