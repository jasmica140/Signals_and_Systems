```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
#1.1
def a(n):
    if (n%2)==0:
        return 0
    else:
        return 4/((math.pi**2)*(n**2))
```

```python
#1.2
plt.figure()

for x in range(-10,11):
    plt.xlabel('n')
    plt.ylabel('a(n)')
    plt.stem(x,a(x),markerfmt=".")
```

=sum(a(k)*(exp(j*k*omeg0*t)+exp(-j*k*omeg0*t) for k in range(0,nmax))
=sum(a(k)*math.cos(t*k*omega0) for k in range(0,nmax))

```python
#1.4
def fhat(t, omega0, nmax):
    
    s = [0]*len(t)
    
    for i in range(0,len(t)):
        s[i] = sum(a(k)*math.cos(t[i]*k*omega0) for k in range(0,nmax+1))

    return s
```

```python
#1.5
nmax = [1,3,5,7,9]
omega0 = 2*math.pi/4
t = np.linspace(-2, 2, 100)

plt.figure()
plt.xlabel('t')
plt.ylabel('truncated Fourier sythesis')
for i in range(0,len(nmax)):
    plt.plot(t,fhat(t, omega0, nmax[i]),label='$nmax = {i}$'.format(i=nmax[i]))

plt.legend(loc='best')
plt.show()
```

```python
#2.1
def f(t):
    
    res = [0]*len(t)

    for i in range(0,len(t)):
        
        if 0<=t[i]<=2:
            res[i] = 1-t[i]
        elif -2<=t[i]<=0:
            res[i] = 1+t[i]
        
    return res
    
```

```python
#2.2
nmax = [1,3,5,7,9]
omega0 = 2*math.pi/4

t = np.linspace(-2, 2, 100)
error = [0]*len(t)

plt.figure()
plt.xlabel('t')
plt.ylabel('error')

for i in range(0,len(nmax)):
    for j in range(0,len(t)):
        error[j] = f(t)[j] - fhat(t, omega0, nmax[i])[j]
    plt.plot(t,error,label='$nmax = {i}$'.format(i=nmax[i]))

plt.legend(loc='best')
plt.show()
```

```python
def E(nmax):
    t = np.linspace(-2, 2, 100)
    omega0 = 2*math.pi/4

    for j in range(0,len(t)):
        error[j] = (f(t)[j] - fhat(t, omega0, nmax)[j])**2
        
    return np.trapz(error)
```

```python
plt.figure()
plt.xlabel('nmax')
plt.ylabel('energy')

for nmax in range(1,21):
    plt.plot(nmax,E(nmax),marker='.',color='b')
```
